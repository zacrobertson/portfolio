# Sentiment Analysis

```python
sentiment = []

for speech in filtered_data.text:
    blob = TextBlob(speech)
    sentiment.append(blob.sentiment.polarity)
```

```python
filtered_data['sentiment'] = sentiment
```

```python
filtered_data[(filtered_data.sentiment > 0) & (filtered_data.show == 'Hannity')]
```
